```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Load data
path = '/content/drive/MyDrive/MSBA 325/leb immigrants.csv'
df = pd.read_csv(path)
```

```python
# Convert numeric columns to proper dtypes
numeric_cols = ['Number of Bangladeshi', 'Number of Sri Lankan', 'Number of Sudanese',
                'Number of Egyptian', 'Number of Ethiopians', 'Number of Iraqi']

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

# Extract district names from URI (assuming it's the last part after the last slash)
df['District'] = df['District URI'].str.split('/').str[-1].str.replace('_', ' ')

# Calculate total immigrants per district
df['Total_Immigrants'] = df[numeric_cols].sum(axis=1)
```

## 1. Bar Chart - Total immigrants by district

```python
fig1 = px.bar(df.sort_values('Total_Immigrants', ascending=False).head(15),
              x='District',
              y='Total_Immigrants',
              title="Total Immigration by District",
              labels={'Total_Immigrants': 'Total Number of Immigrants'})
fig1.update_xaxes(tickangle=45)
fig1.show()
```

## 2. Stacked Bar Chart - Immigration composition by district


```python
top_districts = df.nlargest(10, 'Total_Immigrants')
fig2 = go.Figure()

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']
for i, col in enumerate(numeric_cols):
    if col in df.columns:
        fig2.add_trace(go.Bar(
            name=col.replace('Number of ', ''),
            x=top_districts['District'],
            y=top_districts[col],
            marker_color=colors[i % len(colors)]
        ))

fig2.update_layout(
    title='Immigration Composition by District (Top 10)',
    barmode='stack',
    xaxis_title='District',
    yaxis_title='Number of Immigrants',
    xaxis_tickangle=45
)
fig2.show()
```

## 3. Pie Chart - Overall Distribution by Nationality Across Lebanon

```python
nationality_totals = {}
for col in numeric_cols:
    if col in df.columns:
        total = df[col].sum()
        if total > 0:  # Only include nationalities with immigrants
            nationality_totals[col.replace('Number of ', '')] = total

fig3 = px.pie(values=list(nationality_totals.values()),
              names=list(nationality_totals.keys()),
              title="Total Immigration Distribution by Nationality - Lebanon")
fig3.show()
```

## 4. Scatter Plot - Bangladeshi vs Sri Lankan immigration by district


```python
df_filtered = df[df['Total_Immigrants'] > 0]  # Only districts with immigrants
fig4 = px.scatter(df_filtered,
                  x='Number of Bangladeshi',
                  y='Number of Sri Lankan',
                  size='Total_Immigrants',
                  hover_name='District',
                  title="Bangladeshi vs Sri Lankan Immigration by District",
                  labels={'Number of Bangladeshi': 'Bangladeshi Immigrants',
                         'Number of Sri Lankan': 'Sri Lankan Immigrants'},
                  hover_data=['Number of Sudanese', 'Number of Egyptian'])
fig4.show()
```

## 5. Heatmap - Immigration Across Districts and Nationalities

```python
# Get top 15 districts with most immigrants
top_15 = df.nlargest(15, 'Total_Immigrants')
heatmap_data = top_15[numeric_cols]

fig5 = px.imshow(heatmap_data.values,
                 labels=dict(x="Nationality", y="District", color="Number of Immigrants"),
                 x=[col.replace('Number of ', '') for col in numeric_cols],
                 y=top_15['District'].values,
                 title="Immigration Heatmap - Top 15 Districts",
                 color_continuous_scale='Blues',
                 aspect='auto')
fig5.update_layout(height=600)
fig5.show()
```

## 6. Horizontal Bar Chart

```python
# Extract governorate from Governorate URI
df['Governorate'] = df['Governorate URI'].str.split('/').str[-1].str.replace('_', ' ')

# Group by governorate and sum
gov_data = df.groupby('Governorate')[numeric_cols].sum().reset_index()
gov_data['Total'] = gov_data[numeric_cols].sum(axis=1)
gov_data = gov_data.sort_values('Total', ascending=False)

# Horizontal Stacked Bar for easier reading
fig6 = px.bar(gov_data.sort_values('Total', ascending=True),
              x='Total',
              y='Governorate',
              orientation='h',
              title='Total Immigration by Governorate')
fig6.show()
```
